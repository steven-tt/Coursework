# Important python things

```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

def Lorrenz(xx,t,sigma,rho,beta):
    x=xx[0]
    y=xx[1]
    z=xx[2]
    xp=sigma * (y-x)
    yp=x * (rho-z) - y
    zp= x * y - beta * z
    return np.array([xp,yp,zp])
```

# Parameters and variables and initial conditions

```python
# Chaotic Parameters
sigma = 10.
beta = 8./3.
rho= 28.
rho1= 18.
# Things for graph
t=np.linspace(0,50,10000)
soln1=odeint(Lorrenz, [1,-1,1], t, (sigma,rho1,beta))
soln2=odeint(Lorrenz, [1.001,-1,1], t, (sigma,rho1,beta))
soln3=odeint(Lorrenz, [1,-1,1], t, (sigma,rho,beta))
soln4=odeint(Lorrenz, [1.001,-1,1], t, (sigma,rho,beta))
# Distance
distchaos=np.sqrt((soln3[:,0]-soln4[:,0])**2 +((soln3[:,1]-soln4[:,1]))**2+((soln3[:,2]-soln4[:,2]))**2)
distnochaos=np.sqrt((soln1[:,0]-soln2[:,0])**2 +((soln1[:,1]-soln2[:,1]))**2+((soln1[:,2]-soln2[:,2]))**2)
# print dist
```

# Figures

```python
plt.figure()
plt.subplots_adjust(hspace=.3)
plt.suptitle(r'Lorenz Attractor with $\rho=28$', size=16)

plt.subplot(3,1,1)
plt.ylabel('x solutions')
plt.plot(t,soln3[:,0], color='red', label='$x(0)=1$')
plt.plot(t,soln4[:,0], color='blue', label='$x(0)=1.001$')
plt.legend()

plt.subplot(3,1,2)
plt.ylabel('y solutions')
plt.plot(t,soln3[:,1], color='red')
plt.plot(t,soln4[:,1], color='blue')

plt.subplot(3,1,3)
plt.ylabel('z solutions')
plt.xlabel('t')
plt.plot(t,soln3[:,2], color='red')
plt.plot(t,soln4[:,2], color='blue')

plt.savefig('3_graphs_chaos')
plt.show()
plt.close()

plt.figure()
plt.subplots_adjust(hspace=.3)
plt.suptitle(r'Lorenz Attractor with $\rho=18$', size=16)

plt.subplot(3,1,1)
plt.ylabel('x solutions')
plt.plot(t,soln1[:,0], color='red', label='$x(0)=1$')
plt.plot(t,soln2[:,0], color='blue', label='$x(0)=1.001$')
plt.legend()

plt.subplot(3,1,2)
plt.ylabel('y solutions')
plt.plot(t,soln1[:,1], color='red')
plt.plot(t,soln2[:,1], color='blue')

plt.subplot(3,1,3)
plt.ylabel('z solutions')
plt.xlabel('t')
plt.plot(t,soln1[:,2], color='red')
plt.plot(t,soln2[:,2], color='blue')

plt.savefig('3_graphs_nochaos')
plt.show()
plt.close()

plt.figure()
plt.title(r'Distance between two solutions to the Lorenz System $\rho=28$')
plt.ylabel('Distance between solutions')
plt.xlabel('t')
plt.plot(t,distchaos, color='orange')
plt.savefig('distchaos.png')
plt.show()
plt.close()

plt.figure()
plt.title(r'Distance between two solutions to the Lorenz System $\rho=18$')
plt.ylabel('Distance between solutions')
plt.xlabel('t')
plt.plot(t,distnochaos, color='purple')
plt.plot(0,0,'o', label='Starting point', color='green')
plt.legend()
plt.savefig('distnochaos.png')
plt.show()
plt.close()

fig1=plt.figure()
ax = fig1.gca(projection='3d')
plt.title(r'Lorenz attractor with $\rho=28$')
ax.plot(soln3[:,0],soln3[:,1],soln3[:,2],label='$x(0)=1 $')
ax.plot(soln4[:,0],soln4[:,1],soln4[:,2],label='$x(0)=1.001 $')
plt.savefig('3dchaos.png')
plt.show()
plt.close()

fig2=plt.figure()
ax = fig2.gca(projection='3d')
plt.title(r'Lorenz attractor with $\rho=18$')
ax.plot(soln1[:,0],soln1[:,1],soln1[:,2],label='$x(0)=1 $')
ax.plot(soln2[:,0],soln2[:,1],soln2[:,2],label='$x(0)=1.001 $')
plt.savefig('3dnochaos.png')
plt.show()
plt.close()
```
